```python
import numpy as np
import pandas as pd
from glob import glob


np.random.seed(42)
```

```python
df1 = pd.read_csv("NumtaDB_with_aug/training-a.csv")
df2 = pd.read_csv("NumtaDB_with_aug/training-b.csv")
df3 = pd.read_csv("NumtaDB_with_aug/training-c.csv")
df4 = pd.read_csv("NumtaDB_with_aug/training-d.csv")

df = pd.concat([df1, df2, df3], ignore_index=True)
df.head()
```

```python
df['split_col'] = df['database name original'] + '_' + df['digit'].astype(str)
df.head()
```

```python
df.groupby("split_col").count()
```

```python
# shuffle rows
df = df.sample(frac=1)
df.head()
```

```python
validation_percentage = 0.2

split_col = df['split_col'].unique().tolist()

train_indexes = []

for cat in split_col:
    indexes = df[df['split_col'] == cat].index.tolist()
    train_indexes.extend(indexes[:int(len(indexes) * (1 - validation_percentage))])

train_df = df.loc[train_indexes]
val_df = df.drop(train_indexes)
len(train_df), len(val_df)
```

```python
train_df.to_csv("train.csv", index=False)
val_df.to_csv("val.csv", index=False)
```
